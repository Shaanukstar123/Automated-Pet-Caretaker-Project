# Setup

```python
import torch
from ultralytics import YOLO
import cv2
import pafy #pip install git+https://github.com/Cupcakus/pafy
```

```python
newmodel = YOLO("src/best.pt")
# Test
results = newmodel.predict(source="src/cat.jpg", stream=True)  # save plotted images
```

# Prediction

```python
for result in results:
    for c in result.boxes.cls:
        # Can check this to find cat/dog
        detected = newmodel.names[int(c)]
        print(detected)
        if (detected == "dog"):
            print("Found dog!")
        elif (detected == "cat"):
            print("Found cat!")
        else:
            print("Found something else!")
```

```python
#Youtube streaming
url = "https://www.youtube.com/watch?v=DXUAyRRkI6k"
video = pafy.new(url)
best = video.getbest()

cap = cv2.VideoCapture(best.url)

counter = 0
readeveryn = 1
while True:
    counter += 1
    #Grab frame
    ret, frame = cap.read()
    if not ret:
        break
    #Skip frames
    if counter % readeveryn != 0:
        continue
    %matplotlib inline
    cv2.imshow('frame', frame)
    #Analyse
    results = newmodel.predict(source=frame, stream=True)
    #Get result
    for result in results:
        for c in result.boxes.cls:
            # Can check this to find cat/dog
            detected = newmodel.names[int(c)]
            print(detected)
#must do this
cap.release()
cv2.destroyAllWindows()
```
